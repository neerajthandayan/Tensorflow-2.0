```python
# Importing Libraries

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import tensorflow as tf
```

```python
# Creating the Dataset

N = 1500
X = np.random.random((N,2)) * 5 - 2.5
Y = np.sin(2*X[:,0]) + np.sin(2*X[:,1])
```

```python
# Ploting data

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)
```

```python
# Building Model

reg = tf.keras.models.Sequential()
reg.add(tf.keras.layers.Dense(128, input_shape=(2,), activation=tf.keras.layers.LeakyReLU()))
reg.add(tf.keras.layers.Dense(1))
```

```python
# Compiling Model

optimizer = tf.keras.optimizers.Adam(0.01)
reg.compile(optimizer=optimizer, loss='mse')
```

```python
# Fitting Model

result = reg.fit(X, Y, epochs=100)
```

```python
# Plotting Loss

plt.plot(result.history['loss'], label='Loss')
```

```python
# Plot the prediction surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)

# surface plot
line = np.linspace(-5, 5, 50)
xx, yy = np.meshgrid(line, line)
Xgrid = np.vstack((xx.flatten(), yy.flatten())).T
Yhat = reg.predict(Xgrid).flatten()
ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], Yhat, linewidth=0.2, antialiased=True)
```
